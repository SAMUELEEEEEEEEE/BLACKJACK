```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import joblib
import ast #Per estrapolare una lista da una stringa
import numpy as np
```

```python
data = pd.read_csv("blackjack_simulator.csv")
```

```python
data
```

```python
data_copied = data.copy()
```

```python
data.info()
```

```python
couple_cards_dict = {tuple(ast.literal_eval(cards)) : count for count, cards in enumerate(data["initial_hand"].unique())}
```

```python
actions_dict = {action : count for count, action in enumerate(data["actions_taken"].unique())}
```

```python
initial_hand_numbered = np.array([couple_cards_dict[tuple(ast.literal_eval(cards))] for cards in data["initial_hand"].values])
```

```python
actions_numbered = np.array([actions_dict[action] for action in data["actions_taken"].values])
```

```python
data["initial_hand_numbered"] = initial_hand_numbered
data["actions_numbered"] = actions_numbered
```

```python
data
```

```python
feature = ["dealer_up", "initial_hand_numbered", "run_count"]
target = ["actions_numbered"]
```

```python
X = data[feature].values
y = data[target].values
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.score(X, y)
```
